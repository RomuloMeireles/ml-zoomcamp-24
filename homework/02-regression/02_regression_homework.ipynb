# QUESTION 3

```python
def train_linear_regression(X, y):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])
    
    XTX = X.T.dot(X) # Gram Matrix
    XTX_inv = np.linalg.inv(XTX)
    
    w_full = XTX_inv.dot(X.T).dot(y)
    return w_full[0], w_full[1:] #Bias term and weights

def rmse(y, y_pred):
    se = (y - y_pred) ** 2
    mse = se.mean()
    return np.round(np.sqrt(mse), 2)

base = ['ram', 'storage', 'screen',]
X_train = df_train[base].values

############################################
# Evaluate the Model with Training and Validation Dataset
# Fill in the missing values with 0
X_train_zeros = df_train[base].fillna(0).values
w0, w = train_linear_regression(X_train_zeros, y_train)

y_pred_zeros = w0 + X_train_zeros.dot(w)
rmse_zeros = rmse(y_train, y_pred_zeros)
print(f'training dataset rmse with zeros: {rmse_zeros}')

X_val_zeros = df_val[base].fillna(0).values
val_y_pred_zeros = w0 + X_val_zeros.dot(w)
val_rmse_zeros = rmse(y_val, val_y_pred_zeros)
print(f'validation dataset rmse with zeros: {val_rmse_zeros}')

# Fill in the missing values with mean
mean_screen = df_train['screen'].mean()
X_train_mean = df_train[base].fillna(mean_screen).values

w0_mean, w_mean = train_linear_regression(X_train_mean, y_train)

y_pred_mean = w0_mean + X_train_mean.dot(w_mean)
rmse_mean = rmse(y_train, y_pred_mean)
print(f'training dataset rmse with mean: {rmse_mean}')

X_val_mean = df_val[base].fillna(mean_screen).values
y_pred_val_mean = w0_mean + X_val_mean.dot(w_mean)
rmse_val_mean = rmse(y_val, y_pred_val_mean)
print(f'validation dataset rmse with mean: {rmse_val_mean}')

```

```python
def train_linear_regression_reg(X, y, r=0.001):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])
    
    XTX = X.T.dot(X) # Gram Matrix
    XTX = XTX + r * np.eye(XTX.shape[0])
    
    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)

    return w_full[0], w_full[1:]
```

```python
for r in [0, 0.01, 0.1, 1, 5, 10, 100]:
    X_train = df_train[base].fillna(0).values
    w0, w = train_linear_regression_reg(X_train, y_train, r=r)
    
    X_val = df_val[base].fillna(0).values
    y_pred = w0 + X_val.dot(w)
    
    score = rmse(y_val, y_pred)
    print(r, w0, score)
```

# QUESTION 5

```python
def split_data(df, seed):
    n = len(df)
    n_train = int(n * 0.6)
    n_val = int(n * 0.2)
    n_test = int(n * 0.2)
    
    n_train = n - n_val - n_test

    idx = np.arange(n)
    np.random.seed(seed)
    np.random.shuffle(idx)

    df_train = df.iloc[idx[:n_train]]
    df_val = df.iloc[idx[n_train:n_train + n_val]]
    df_test = df.iloc[idx[n_train + n_val:]]

    df_train = df_train.reset_index(drop=True)
    df_val = df_val.reset_index(drop=True)
    df_test = df_test.reset_index(drop=True)

    y_train = df_train.final_price.values
    y_val = df_val.final_price.values
    y_test = df_test.final_price.values

    df_train = df_train.drop(columns=['final_price'])
    df_val = df_val.drop(columns=['final_price'])
    df_test = df_test.drop(columns=['final_price'])

    return df_train, df_val, df_test, y_train, y_val, y_test
```

```python

def evaluate_model_on_seeds(df, seeds, base_features):
    rmse_scores = []

    for seed in seeds:
        df_train, df_val, df_test, y_train, y_val, y_test = split_data(df, seed=seed)
        
        X_train = df_train[base_features].fillna(0).values
        X_val = df_val[base_features].fillna(0).values
        
        w0, w = train_linear_regression(X_train, y_train)

        y_val_pred = w0 + X_val.dot(w)

        score = rmse(y_val, y_val_pred)

        rmse_scores.append((seed, score))
    
    return rmse_scores
```

```python
def evaluate_model_on_seeds_reg(df, seed, base_features, r=0.001):

    df_train, df_val, df_test, y_train, y_val, y_test = split_data(df, seed)
    
    df_combined = pd.concat([df_train, df_val])
    y_combined = np.concatenate([y_train, y_val])
    
    X_combined = df_combined[base_features].fillna(0).values
    X_test = df_test[base_features].fillna(0).values
    
    w0, w = train_linear_regression_reg(X_combined, y_combined, r)

    y_test_pred = w0 + X_test.dot(w)

    score = rmse(y_test, y_test_pred)
    
    return score
```
